```python
def avg_valence_new(data: pl.DataFrame,
                    lexicon: pl.DataFrame,
                    backbone: str = "spacy",
                    language: str = "en",
                    word_column: str = "word",
                    ) -> float:
    lex_lookup = lexicon.select(pl.col(word_column).alias("lemma"), pl.col("valence"))
    tmp = data.with_row_count("row_id").explode("lemmas")
    means = tmp.join(lex_lookup, left_on="lemmas", right_on="lemma", how="left") \
            .group_by("row_id").agg(pl.col("valence").mean().alias("avg_valence_new"))
    data = data.join(means, on="row_id", how="left").drop("row_id")
    return data
```
